```python
def func(num):
    summ=0
    mystr=str(num)
    for digit in mystr:
        summ+=int(digit)
    if summ==10:
        return True
    else:
        return False
```

```python
li=[]
count=0
for i in range(10000):
    if func(i)==True:
        li.append(i)
        count+=1
print(li)
print(count)
li2=[]
for i in range(len(li)-1):
    li2.append(li[i+1]-li[i])
print(li2)
```
